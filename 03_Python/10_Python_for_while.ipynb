# 반복문 for, while
* 반복문은 동일한 작업을 여러번 반복해야할 경우 사용
* for, while 두 종류의 반복문이 있음
* for : 반복 횟수가 정해져 있는 경우
* while : 반복 횟수가 정해져 있지 않은 경우 (무한 반복)

## for 반복문
* 반복할 횟수가 정해져 있는 경우

for 변수1, 변수2.. in 반복할 객체/자료: <br>
____ 반복할 코드

* 순서가 있는 자료형(문자열, 리스트, 튜플) + 순서는 없지만 dict도 반복문에서 사용 가능

```python
for i in range(1,11):
    print(i)
```

```python
# 문자열
for char in '파이썬은 문자열을 반복문에 넣어 출력 가능':
    print(char, end=' ')
```

```python
# 리스트
for item in [1,2,3,'사','오','육',[7,8,9],(10,11),{'key':'value'}]:
    print(item)
```

```python
# 딕셔너리
data = dict(이름='홍길동',나이=40, 주소='서울',전화번호='010-0000-0000')
```

```python
for item in data:
    print(item)
    
print()

for item in data.keys():
    print(item)
```

```python
for item in data:
    print(data[item])
    
print()

for item in data.values():
    print(item)
```

```python
for key, value in data.items():
    print(key, value)
```

* for 반복문을 이용해서 1부터 10까지 정수를 더한 합을 출력

```python
sum = 0

for i in range(1,11):
    sum += i
    print(sum)
```

## enumerate()
* 반복문에서 자료의 인덱스를 만들어주는 함수

```python
s = 'enumerate 함수는 반복문에서 index를 만들어 출력해주는 함수입니다.'

import time
for idx, i in enumerate(s):
    print(f'반복  : {idx+1}/{len(s)}', f'값 : {i}',end='\r')
    time.sleep(0.5)
```

## continue, break
* continue : 코드 실행을 건너뛰고 반복을 진행
* break : 특정 조건이 되었을 때 반복문 정지

```python
for idx, i in enumerate(s):
    if (idx+1) %2 == 0:
        continue
    print(f'반복  : {idx+1}/{len(s)}', f'값 : {i}',end='\r')
    time.sleep(0.5)
```

```python
for idx, i in enumerate(s):
    print(f'반복  : {idx+1}/{len(s)}', f'값 : {i}',end='\r')
    if (idx+1) == 10:
        break
    time.sleep(0.5)
```

# while 반복문
* 반복할 횟수가 정해지지 않은 경우
* 무한 반복 가능
* 무한 반복이 되지 않도록 조건식을 잘 짜주는 것이 중요

변수 선언 및 초기화
while 조건식: <br>
____ 반복할 코드 <br>
____ 변수 증가/감소 코드 <br>
____ 반복문이 종료될 조건 + break

```python
num = 1
while(num < 11):
    print(num, end=' ')
    num += 1
```

* 1-50까지의 수 중에서 3의 배수이거나, 짝수인 수를 찾아서 더한 값을 출력

```python
sum1 = 0
for i in range(1,51):
    if i %3 == 0 or i %2 == 0:
        sum1 += i
print(sum1)
```

## while의 무한 반복
* while 뒤의 조건이 True인 경우 무한 반복
* while문 내에서 if 조건문으로 break를 걸어주어야 멈춤

```python
iterate1 = 0

while True:
    iterate1 = int(input('정수 : '))
    if iterate1 > 100:
        break
```

* 무한 반복이 되는 조건 : True
* True가 되는 경우 : 1, 정수, 실수, 문자열, list, tuple, dict, True
* False가 되는 경우 : 0, 0.0, "", '', [], (), {}, False

# 중첩 반복문
* 반복문을 2개 이상 겹쳐서 사용
* 바늘 시계의 원리와 같음
* 안쪽 반복문이 끝난 후 바깥쪽 반복문이 실행
* 초침 → 분침 → 시침

* 구구단 만들기 2중 중첩 for 문
* 2단 ~ 9단, x1 ~ x9

```python
import time
for i in range(1,10):
    for j in range(1,10):
        print(f'{i}x{j} = {i*j}', end='\r')
        time.sleep(0.2)
```

* 반복문 3개로 시계 만들기
* 00 - 23 시
* 00 - 59 분
* 00 - 59 초

```python
for i in range(0, 24):
    for j in range(0, 60):
        for z in range(0, 60):
            print(f'{i:02d}시 {j:02d}분 {z:02d}초', end='\r')
#             time.sleep(0.0001)
```

## 중첩 반복문을 사용해서 다중 리스트 출력하기

```python
l1 = []
for i in range(1,6):
    l1.append(list(range(1,6)))
l1
```

```python
for i in l1:
    for j in i:
        print(j, end=' ')
    print()
```

```python
animal = ['dog', 'cat', 'parrot']
for val in animal:
    print(val[0])
```

```python
l2 = ['l', 'study', 'python', 'language', '!']
for i in l2:
    if len(i) >= 3:
        print(i)
```

```python
apart = [[101, 102], [201, 202], [301, 302]]

for i in apart:
    for j in i:
        print(f'{j}호', end = ' ')
    print()
```

```python
for i in sorted(apart, reverse=True):
    for j in i:
        print(f'{j}호', end = ' ')
    print()
```

```python
for i in apart[::-1]:
    for j in i:
        print(f'{j}호', end = ' ')
    print()
```
