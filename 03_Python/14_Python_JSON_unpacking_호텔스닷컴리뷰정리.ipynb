```python
import json
import pandas as pd
```

```python
with open('./data/hotels_korea_reviews_text.json','r',encoding='utf-8') as file:
    temp = file.read()
    data = json.loads(temp)
print(len(data))
```

```python
hotel_info = {}
users_reviews = {}

for idx, d in enumerate(data):
    hotel_info.setdefault('hotelID',[]).append(d['data']['body']['query']['destination']['id'])
    id = d['data']['body']['query']['destination']['id']
    
    hotel_info.setdefault('호텔이름',[]).append(d['data']['body']['propertyDescription']['name'])
    hotel_info.setdefault('호텔등급',[]).append(d['data']['body']['propertyDescription']['starRatingTitle'])
    hotel_info.setdefault('총리뷰수',[]).append(d['data']['body']['reviewContent']['overall']['totalCount'])
    hotel_info.setdefault('평점',[]).append(d['data']['body']['reviewContent']['overall']['rating'])
    
    #     for key, value in d['data']['body']['reviewContent']['overall']['ratingAspects'].items():
    #          hotel_info.setdefault( key,[]).append(value)
    hotel_info.setdefault('청결도',[]).append(d['data']['body']['reviewContent']['overall']['ratingAspects'].get('cleanliness','없음'))
    hotel_info.setdefault('서비스',[]).append(d['data']['body']['reviewContent']['overall']['ratingAspects'].get('service','없음'))
    hotel_info.setdefault('편안함',[]).append(d['data']['body']['reviewContent']['overall']['ratingAspects'].get('comfort','없음'))
    hotel_info.setdefault('시설상태',[]).append(d['data']['body']['reviewContent']['overall']['ratingAspects'].get('condition','없음'))
    hotel_info.setdefault('주변환경',[]).append(d['data']['body']['reviewContent']['overall']['ratingAspects'].get('neighbourhood','없음'))

    hotel_info.setdefault('지역',[]).append(d['data']['body']['propertyDescription']['address'].get('region','없음'))
    hotel_info.setdefault('시도',[]).append(d['data']['body']['propertyDescription']['address']['locality'])
    
    try:
        for rv in d['data']['body']['reviewContent']['reviews']['hermes']['groups'][0]['items']:
            users_reviews.setdefault('hotelID',[]).append(id)
            users_reviews.setdefault('리뷰일',[]).append(rv['reviewDate'])
            users_reviews.setdefault('고객id',[]).append(rv['itineraryId'])
            users_reviews.setdefault('여행타입',[]).append(rv['tripType'])
            users_reviews.setdefault('평점',[]).append(rv['rating'])
            users_reviews.setdefault('리뷰',[]).append(rv['description'])
    except Exception as e:
        print(e)
        print(f'error idx : {idx}')

print(len(hotel_info))
print(hotel_info.keys())
for key in hotel_info:
    print(len(hotel_info.get(key)))
```

```python
print(len(users_reviews))
print(users_reviews.keys())
for key in users_reviews:
    print(len(users_reviews.get(key)))
```
