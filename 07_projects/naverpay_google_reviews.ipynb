```python
from google_play_scraper import reviews, Sort
from datetime import datetime
import time

naverpay_reviews = []
continuation_token = None

# 수집 중단 기준 날짜 설정
cutoff_date = datetime(2025, 4, 30)

while True:
    result_log, continuation_token = reviews(
        'com.naverfin.payapp',
        lang='ko',
        country='kr',
        sort=Sort.NEWEST,
        count=100,
        continuation_token=continuation_token
    )

    # 날짜 조건을 걸어서 중단 여부 판단
    for review in result_log:
        if review['at'] < cutoff_date:
            continuation_token = None
            break
        naverpay_reviews.append(review)

    if continuation_token is None:
        break

    time.sleep(1)  # 과도한 요청 방지용
```

```python
naverpay_reviews
```

```python
naverpay_df = pd.DataFrame(naverpay_reviews)
```
