```python
page = 0
chicken = {}

while True:
    page += 1
    payload = dict(servicekey='lk+57ADP0ai9JW/Do7IUnAdPT24P399mqf6XY6BiyIumPIGX1Hpvad7OeAQQaPCgNmTetZjAy8AreL6zI9Q3ag==',
              divId='indsSclsCd', key='I21006', numOfRows=1000, pageNo=page, type='json' )
    req = requests.get(url, params=payload)
    ch_data = req.json()
    print(page, ':', len(ch_data['body']['items']), end='\r')
    if len(ch_data['body']['items']) == 0:
        break
    for item in ch_data['body']['items']:
        for col, val in zip(ch_data['header']['columns'], item.values()):
            chicken.setdefault(col,[]).append(val)
```
