```python
page = 0
result = {}

while True:
    page += 1
    payload2 = dict(serviceKey = 'lk+57ADP0ai9JW/Do7IUnAdPT24P399mqf6XY6BiyIumPIGX1Hpvad7OeAQQaPCgNmTetZjAy8AreL6zI9Q3ag==',
              numOfRows=10, pageNo=page, dataType='JSON')
    req = requests.get(url, params=payload2)
    d = req.json()
    print(page,':',len(d['body']['items']))
    if len(d['body']['items']) == 0:
        break
    for item in d['body']['items']:
        for key, val in item.items():
            result.setdefault(key,[]).append(val)
```

```python
print(result)
```

```python
import pandas as pd

result_df = pd.DataFrame(result)
result_df
```

# BeautifulSoup
* https://www.crummy.com/software/BeautifulSoup/bs4/doc/# 참고
* pip install beautifulsoup4
* from bs4 import BeautifulSoup as bs
* find, find_all 함수 : xml, html에서 태그 기반으로 내용을 찾음
* select, select_one 함수 : xml, html에서 css selector 기반으로 내용을 찾음
* find_all, select는 해당 태그나 css selector를 가진 부분을 모두 찾아서 list로 반환
* find, select_one은 여러 태그나 css selector 중에서 가장 먼저 나오는 태그/selector를 한 개만 찾아줌
* 찾아온 태그 .name : 태그 이름 반환
* 찾아온 태그 .text, .string : 태그 안쪽의 텍스트 반환

```python
new_url = 'https://apis.data.go.kr/1160100/service/GetSmallLoanFinanceInstituteInfoService/getOrdinaryFinanceInfo'
```

```python
new_test = dict(serviceKey = 'lk+57ADP0ai9JW/Do7IUnAdPT24P399mqf6XY6BiyIumPIGX1Hpvad7OeAQQaPCgNmTetZjAy8AreL6zI9Q3ag==',
               numOfRows=500, pageNo=11, resultType='xml')
```

```python
new_req = requests.get(new_url, params=new_test)
```

```python
print(new_req.url)
print(new_req.status_code)
```

```python
print(new_req.text)
```

```python
new_data = bs(new_req.text,'xml')
new_data.select('item')
```

```python
for item in new_data.select('item'):
    for tag in item:
        print(tag.name, tag.text)
```

```python
import time

page = 0
ord_result_xml = {}

while True:
    page += 1
    ord_xml_payload = dict(serviceKey = 'lk+57ADP0ai9JW/Do7IUnAdPT24P399mqf6XY6BiyIumPIGX1Hpvad7OeAQQaPCgNmTetZjAy8AreL6zI9Q3ag==',
               numOfRows=500, pageNo=page, resultType='xml')
    ord_xml_req = requests.get(new_url, params=ord_xml_payload)
    xml_data = bs(ord_xml_req.text, 'xml')
    print(page, ':', len(xml_data.select('item')), end='\r')
    if len(xml_data.select('item')) == 0:
        break
    for item in xml_data.select('item'):
        for tag in item:
            if tag.name != None:
                ord_result_xml.setdefault(tag.name,[]).append(tag.text)
    time.sleep(0.2)
```

```python
print(ord_result_xml.keys())
```

```python
ord_xml_df = pd.DataFrame(ord_result_xml)
ord_xml_df
```

```python
page = 0
ord_result_json = {}

while True:
    page += 1
    ord_json_payload = dict(serviceKey = 'lk+57ADP0ai9JW/Do7IUnAdPT24P399mqf6XY6BiyIumPIGX1Hpvad7OeAQQaPCgNmTetZjAy8AreL6zI9Q3ag==',
               numOfRows=500, pageNo=page, resultType='json')
    ord_json_req = requests.get(new_url, params=ord_json_payload)
    json_data = ord_json_req.json()
    if len(json_data['response']['body']['items']['item']) == 0:
        break
    print(page, ':', len(json_data['response']['body']['items']['item']), end='\r')
    for item in json_data['response']['body']['items']['item']:
        for key, val in item.items():
            ord_result_json.setdefault(key, []).append(val)
```

```python
ord_json_df = pd.DataFrame(ord_result_json)
ord_json_df
```
